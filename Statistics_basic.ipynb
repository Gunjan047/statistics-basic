**Question 1: What is the difference between descriptive statistics and inferential statistics?          Explain with examples.**

 **Answer**

**1. Descriptive Statistics**

 - **Definition:**
  Descriptive statistics summarize and describe the main features of a dataset. They focus only on the data at hand, without making predictions or generalizations beyond it.

**Purpose:**

 To present data in a meaningful way so patterns can be seen easily.

**Examples of methods:**


 - **Measures of central tendency:** Mean, Median, Mode

 - **Measures of dispersion:** Range, Variance, Standard Deviation

 - **Data visualizations:** Histograms, Pie charts, Boxplots

**Example:**
   - A teacher calculates the average score of her class in a math test (say 72/100).

   - This only describes the students in that class.

   - It does not make any prediction about other classes.

**2. Inferential Statistics**

 - **Definition:**

  Inferential statistics use sample data to make generalizations, predictions, or decisions about a larger population.

 - **Purpose:** To go beyond the immediate data and infer conclusions.

**Examples of methods:**

   - Hypothesis testing (e.g., t-test, chi-square test)

  - Confidence intervals

  - Regression analysis

   - ANOVA (Analysis of Variance)

**Example:**

 - Suppose the teacher takes a sample of 30 students from all schools in a district and calculates their average math score.

 - Using inferential statistics, she might estimate the average score of all students in the district.

 - She could also test: “Is the average score significantly higher in urban schools compared to rural schools?”

**Question 2: What is sampling in statistics? Explain the differences between random and stratified sampling.**

 - In statistics, sampling is the process of selecting a subset (called a sample) from a larger group (called the population) to make inferences or draw conclusions about the population.

  Since studying an entire population is often expensive, time-consuming, or impractical, sampling helps researchers analyze data more efficiently while still obtaining reliable insights.

**Example:**

If a company wants to study customer satisfaction across 10,000 customers, it may survey only 500 customers (the sample) instead of all 10,000.

**Random Sampling**

 - In random sampling, each individual in the population has an equal chance of being selected.

 - This method minimizes selection bias and ensures the sample is representative (on average).

**Example:**

If you randomly select 500 customers from a list of 10,000 using a random number generator, that’s random sampling.

**Advantages:**

- Simple to understand and implement.

 - Reduces bias.

- **Limitations:**

  May not represent smaller subgroups well (e.g., age groups, genders).

**Stratified Sampling**

- In stratified sampling, the population is divided into subgroups (strata) based on shared characteristics (e.g., age, gender, income level).

- Then, random samples are taken from each stratum, often in proportion to their size in the population.

**Example:**

  If the 10,000 customers consist of 60% women and 40% men, and you want a sample of 500, you first divide customers into men and women, then randomly pick 300 women and 200 men.

**Advantages:**

 - Ensures representation of all key subgroups.

 - Produces more precise and reliable results than simple random sampling.

**Limitations:**

 - Requires prior knowledge of population characteristics.

 - More complex to implement.

**Question 3: Define mean, median, and mode. Explain why these measures of central tendency are important.**

  **Answer** - In statistics, the mean, median, and mode are measures of central tendency used to identify a single value that represents the center or typical value of a dataset. They are crucial because they summarize complex data, provide insights into data characteristics, and guide decision-making. The most appropriate measure to use depends on the data's distribution and the goal of the analysis.

**Mean**

  The mean is the arithmetic average of a dataset. Calculation: Sum all the values and divide by the total number of values.

**Example:**

 For the dataset {2,3,3,4,6,8,9}, the mean is:(2+3+3+4+6+8+9)/7=35/7=5).

**Median**

  The median is the middle value of a dataset sorted in order. For an odd number of values, it's the single middle value; for an even number, it's the average of the two middle values. For {2,3,3,4,6,8,9} the median is 4.

 **Mode**

  The mode is the most frequent value in a dataset. A dataset can have no mode, one mode, or multiple modes. In {2,3,3,4,6,8,9}, the mode is 3.

 **Why These Measures Are Important**


- **Summarize Data**

  They provide a single value that represents the "center" of the dataset.

- **Comparison**

  Useful for comparing different datasets. (e.g., average income across countries).

- **Understand Distribution**

    - Mean shows the overall average.

    - Median is better when data has outliers/skewness (e.g., income data where a few very rich people raise the mean).

    - Mode is useful for categorical data (e.g., most common shoe size, favorite color).

- **Decision-Making**

    Businesses, governments, and researchers use them to make data-driven decisions (e.g., setting salaries, pricing policies, medical research).






**Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?**

-  **Skewness**

 **Definition:**

   - Skewness measures the asymmetry of a data distribution around its mean.

   - A perfectly symmetric distribution (like the normal distribution) has skewness = 0.

**Types of Skewness:**

  - Positive Skew (Right-Skewed):

  - Tail is longer on the right side.

  - Mean > Median > Mode.

Example: Income distribution (a few very high earners raise the mean).

**Negative Skew (Left-Skewed):**

 - Tail is longer on the left side.

  - Mean < Median < Mode.

  - Example: Age at retirement (most people retire at ~60, but a few retire earlier).

**2. Kurtosis**

**Definition:**Kurtosis measures the “tailedness” or how heavily the tails of a distribution differ from the normal distribution.

  - A normal distribution has kurtosis = 3 (called mesokurtic).

**Types of Kurtosis:**

  - Leptokurtic (Kurtosis > 3):

    - Heavy tails, more outliers.

    - Distribution is sharper than normal.

**Platykurtic (Kurtosis < 3):**

  - Light tails, fewer outliers.

  - Distribution is flatter than normal.

**Mesokurtic (Kurtosis = 3):**

  - Same tail thickness as a normal distribution.

**What does a Positive Skew imply?**

 **A positive skew means:**

  - The distribution has a long right tail.

  - Most data points are concentrated on the left side (lower values).

  - The mean is pulled to the right by extreme large values.

Example: Monthly income of a population – most people earn moderate amounts, but a few earn extremely high salaries, stretching the tail to the right.


**Question 5: Implement a Python program to compute the mean, median, and mode of
a given list of numbers.
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]
(Include your Python code and output in the code box below.)**

```python
import statistics as stats

# Given list of numbers
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Calculate mean, median, and mode
mean_val = stats.mean(numbers)
median_val = stats.median(numbers)
mode_val = stats.mode(numbers)

# Print results
print("Numbers:", numbers)
print("Mean:", mean_val)
print("Median:", median_val)
print("Mode:", mode_val)
```

**Question 6: Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python: list_x = [10, 20, 30, 40, 50] list_y = [15, 25, 35, 45, 60]**




```python
import numpy as np

# Given datasets
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Convert to numpy arrays
x = np.array(list_x)
y = np.array(list_y)

# Compute covariance matrix
cov_matrix = np.cov(x, y, bias=False)   # bias=False → sample covariance
covariance = cov_matrix[0, 1]

# Compute correlation coefficient
correlation = np.corrcoef(x, y)[0, 1]

# Print results
print("List X:", list_x)
print("List Y:", list_y)
print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

**Question 7: Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result: data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]**

```python
import matplotlib.pyplot as plt
import numpy as np

# Given data
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Create a boxplot
plt.boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor="lightblue"))
plt.title("Boxplot of Data")
plt.xlabel("Values")
plt.show()

# --- Outlier detection using IQR method ---
Q1 = np.percentile(data, 25)  # 1st quartile
Q3 = np.percentile(data, 75)  # 3rd quartile
IQR = Q3 - Q1                 # Interquartile Range

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("Q1 (25th percentile):", Q1)
print("Q3 (75th percentile):", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Outliers:", outliers)
```

**Question 8: You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales. ● Explain how you would use covariance and correlation to explore this relationship. ● Write Python code to compute the correlation between the two lists: advertising_spend = [200, 250, 300, 400, 500] daily_sales = [2200, 2450, 2750, 3200, 4000]**

**Covariance**

  - whether the two variables move in the same direction (positive covariance) or in opposite directions (negative covariance).

**Example:** If ad spend increases and sales also increase → covariance is positive.

**Correlation**

 - Goes further by standardizing covariance so we can measure the strength of the relationship.

Value range: -1 to +1.

 - +1 → perfect positive correlation (sales rise proportionally with spend).

 - 0 → no relationship.

 - -1 → perfect negative correlation.

Correlation is unitless, so it’s easier to interpret than covariance.

```python
import numpy as np

# Given data
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Convert to numpy arrays
x = np.array(advertising_spend)
y = np.array(daily_sales)

# Covariance matrix
cov_matrix = np.cov(x, y, bias=False)
covariance = cov_matrix[0, 1]

# Correlation coefficient
correlation = np.corrcoef(x, y)[0, 1]

# Print results
print("Advertising Spend:", advertising_spend)
print("Daily Sales:", daily_sales)
print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

**Question 9: Your team has collected customer satisfaction survey data on a scale of 1-10 and wants to understand its distribution before launching a new product. ● Explain which summary statistics and visualizations (e.g. mean, standard deviation, histogram) you’d use. ● Write Python code to create a histogram using Matplotlib for the survey data: survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]**

**Summary Statistics**

 - Mean (average) → central tendency of satisfaction.

 - Median → robust measure of central tendency (less affected by outliers).

 - Mode → most common satisfaction score.

 - Standard Deviation / Variance → spread of scores (are most customers clustered or widely varied?).

 - Minimum & Maximum → range of satisfaction values.

**Visualizations**

 - Histogram → shows frequency distribution of scores.

 - Boxplot → highlights median, quartiles, and potential outliers.

 - Bar chart of counts → useful since the scale is discrete (1–10).

```python
import matplotlib.pyplot as plt

# Given survey scores
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Create histogram
plt.hist(survey_scores, bins=range(1, 12), edgecolor='black', color='skyblue', rwidth=0.8)

# Add labels and title
plt.xlabel("Satisfaction Score (1–10)")
plt.ylabel("Frequency")
plt.title("Distribution of Customer Satisfaction Scores")
plt.xticks(range(1, 11))  # ensure ticks from 1–10
plt.show()
```
